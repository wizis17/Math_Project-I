## Principal component Analysis (PCA) using SVD

Though not exactly the same, SVD can be used to perform PCA, which is a widely used technique for dimensionality and data visualization

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import TruncatedSVD
from sklearn.datasets import load_digits

# Step 1: Load high-dimensional dataset (Handwritten Digits Dataset)
digits = load_digits()
X = digits.data  # 64-dimensional feature space

print("Original Data Shape:", X.shape)  # (1797, 64) -> 64D features
```

```python
# Step 2: Apply PCA using SVD (reduce to 2D for visualization)
pca_svd = TruncatedSVD(n_components=2)  # Reduce to 2 components
X_pca = pca_svd.fit_transform(X)

print("Reduced Data Shape:", X_pca.shape)  # (1797, 2) -> 2D features
```

```python
# Step 3: Scatter plot of the reduced data
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=digits.target, cmap='viridis', alpha=0.7)
plt.colorbar(label="Digit Label")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA (SVD-based) - Digits Dataset")
plt.show()
```
