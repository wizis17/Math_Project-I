# Matrix Approximation using SVD

Decompose a large matrix A into three matrices : 
# A=U⋅S⋅VT 

Keep only the top k singular values to approximate the original matrix with lower rank

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import svd

# Step 1: Create a large random matrix (10x10)
np.random.seed(42)
A = np.random.rand(10, 10)  # Original large matrix
print("Original Matrix:\n", A)
```

```python
# Step 2: Apply SVD
U, S, Vt = svd(A)

# Step 3: Approximate using only top k singular values
k = 3  # Reduce to rank-3 approximation
S_k = np.zeros((k, k))
np.fill_diagonal(S_k, S[:k])

A_approx = U[:, :k] @ S_k @ Vt[:k, :]  # Matrix approximation

print("\nApproximated Matrix (Rank-3):\n", A_approx)
```

```python
# Step 4: Visualizing the Difference
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.imshow(A, cmap='viridis')
plt.title("Original Matrix")
plt.colorbar()

plt.subplot(1, 2, 2)
plt.imshow(A_approx, cmap='viridis')
plt.title(f"Rank-{k} Approximation")
plt.colorbar()

plt.show()
```

✔ Step 1: Generate a 10×10 random matrix to represent large data.

✔ Step 2: Apply SVD decomposition to split the matrix into U, S, and Vt.

✔ Step 3: Keep only the top 3 singular values for a low-rank approximation.

✔ Step 4: Compare the original vs approximated matrix visually.
