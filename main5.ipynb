Solving Dimensionality Reduction using SVD

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import TruncatedSVD

# Step 1: Create a sample high-dimensional dataset (6x6 matrix)
np.random.seed(42)
original_matrix = np.random.rand(6, 6)  # Random 6x6 matrix
print("Original Matrix:\n", original_matrix)

```

```python
# Step 2: Apply SVD and reduce dimensions
num_components = 2  # Keep only 2 significant singular values
svd = TruncatedSVD(n_components=num_components)
reduced_matrix = svd.fit_transform(original_matrix)
```

```python
# Step 3: Reconstruct the approximate matrix
approx_matrix = svd.inverse_transform(reduced_matrix)

print("\nReduced Matrix (2 Components):\n", reduced_matrix)
print("\nApproximated Matrix (Reconstructed from 2 Components):\n", approx_matrix)
```

```python

# Step 4: Visualize the Singular Values
plt.plot(svd.singular_values_, marker='o', linestyle='-')
plt.title("Singular Values - Importance of Components")
plt.xlabel("Component Index")
plt.ylabel("Singular Value")
plt.show()
```

✔ Step 1: Generate a 6×6 random matrix to represent high-dimensional data.

✔ Step 2: Apply SVD and keep only 2 singular values (dimensionality reduction).

✔ Step 3: Reconstruct an approximate version of the original matrix.

✔ Step 4: Plot the Singular Values to see their importance.
